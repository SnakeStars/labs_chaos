Исследуется автономная система обыкновенных дифференциальных уравнений:

$$
\begin{cases}
\dot{x} = 2.7y + z \\
\dot{y} = -x + y^2 \\
\dot{z} = x + y
\end{cases}
$$

Необходимо:
1. Найти стационарные точки и исследовать их устойчивость
2. Построить фазовый портрет
3. Построить сечения Пуанкаре
4. Исследовать расхождение близких траекторий
5. Вычислить спектр показателей Ляпунова

## 2. Теоретические основы

### 2.1 Стационарные точки и устойчивость

Стационарные (равновесные) точки системы находятся из условий $\dot{x}=0$, $\dot{y}=0$, $\dot{z}=0$:

$$
\begin{cases}
2.7y + z = 0 \\
-x + y^2 = 0 \\
x + y = 0
\end{cases}
$$

Из третьего уравнения $x = -y$. Подставляя во второе: $-(-y) + y^2 = y + y^2 = y(1+y) = 0$, откуда $y = 0$ или $y = -1$.

Получаем две стационарные точки:
- $P_1 = (0, 0, 0)$
- $P_2 = (1, -1, 2.7)$

### 2.2 Исследование устойчивости по первому приближению

Устойчивость определяется собственными значениями матрицы Якоби:

$$
J = \begin{pmatrix}
\frac{\partial \dot{x}}{\partial x} & \frac{\partial \dot{x}}{\partial y} & \frac{\partial \dot{x}}{\partial z} \\
\frac{\partial \dot{y}}{\partial x} & \frac{\partial \dot{y}}{\partial y} & \frac{\partial \dot{y}}{\partial z} \\
\frac{\partial \dot{z}}{\partial x} & \frac{\partial \dot{z}}{\partial y} & \frac{\partial \dot{z}}{\partial z}
\end{pmatrix} = \begin{pmatrix}
0 & 2.7 & 1 \\
-1 & 2y & 0 \\
1 & 1 & 0
\end{pmatrix}
$$

**Для точки $P_1(0,0,0)$:**
$$
J(P_1) = \begin{pmatrix}
0 & 2.7 & 1 \\
-1 & 0 & 0 \\
1 & 1 & 0
\end{pmatrix}
$$

Характеристическое уравнение:
$$
\det(J - \lambda I) = \begin{vmatrix}
-\lambda & 2.7 & 1 \\
-1 & -\lambda & 0 \\
1 & 1 & -\lambda
\end{vmatrix} = -\lambda^3 - 1.7\lambda - 1 = 0
$$
$$
\lambda^3 + 1.7\lambda + 1 = 0
$$

Корни: $\lambda_1 \approx -0.676$ (вещественный отрицательный), два комплексно-сопряженных с положительной вещественной частью. Следовательно, $P_1$ — **седло-фокус**, неустойчива.

**Для точки $P_2(1,-1,2.7)$:**
При $y = -1$ имеем $2y = -2$:
$$
J(P_2) = \begin{pmatrix}
0 & 2.7 & 1 \\
-1 & -2 & 0 \\
1 & 1 & 0
\end{pmatrix}
$$

Характеристическое уравнение:
$$
\lambda^3 + 2\lambda^2 + 1.7\lambda - 1 = 0
$$

Корни: $\lambda_1 \approx 0.312$ (положительный), $\lambda_{2,3} \approx -1.156 \pm 1.157i$ (комплексные с отрицательной вещественной частью). $P_2$ — также **седло-фокус**.

Обе стационарные точки являются седловыми, что характерно для систем с хаотическим поведением.

## 3. Теория хаотических аттракторов

### 3.1 Определение аттрактора

**Аттрактор** — замкнутое инвариантное множество в фазовом пространстве, которое:
- притягивает траектории из некоторой окрестности (бассейна притяжения)
- не может быть разложено на два непересекающихся замкнутых инвариантных подмножества

**Странный (хаотический) аттрактор** — аттрактор, обладающий:
1. **Чувствительной зависимостью от начальных условий** (положительный показатель Ляпунова)
2. **Фрактальной структурой** (дробная размерность)
3. **Перемешиванием** (эргодичностью)

### 3.2 Показатели Ляпунова

**Показатели Ляпунова** $\lambda_i$ характеризуют среднюю скорость экспоненциального разбегания (или сближения) близких траекторий вдоль линейно независимых направлений в фазовом пространстве.

Для трехмерной системы имеем три показателя:
- $\lambda_1 > 0$ — направление экспоненциального растяжения (хаос)
- $\lambda_2 = 0$ — направление вдоль траектории (движение)
- $\lambda_3 < 0$ — направление экспоненциального сжатия

#### 3.2.1 Метод Бенеттина вычисления показателей Ляпунова

Алгоритм:
1. Интегрируем исходную систему и уравнения в вариациях одновременно
2. Через интервалы времени $T$ проводим ортогонализацию векторов возмущений методом Грама–Шмидта
3. Накопливаем сумму логарифмов норм после ортогонализации
4. Показатели Ляпунова вычисляются как:

$$
\lambda_i = \lim_{N \to \infty} \frac{1}{NT} \sum_{k=1}^{N} \ln \|\tilde{\mathbf{x}}_i^{(k)}\|
$$

#### 3.2.2 Уравнения в вариациях

Для системы $\dot{\mathbf{x}} = \mathbf{f}(\mathbf{x})$ уравнения в вариациях имеют вид:

$$
\delta \dot{\mathbf{x}} = J(\mathbf{x}) \cdot \delta \mathbf{x}
$$

где $J(\mathbf{x})$ — матрица Якоби.

Для нашей системы:
- $\delta\dot{x} = 2.7\,\delta y + \delta z$
- $\delta\dot{y} = -\delta x + 2y\,\delta y$
- $\delta\dot{z} = \delta x + \delta y$

### 3.3 Связь с дивергенцией

**Теорема** (связь показателей Ляпунова с дивергенцией):
$$
\sum_{i=1}^{n} \lambda_i = \lim_{T \to \infty} \frac{1}{T} \int_0^T \operatorname{div} \mathbf{f}(\mathbf{x}(t)) \, dt
$$

Дивергенция векторного поля:
$$
\operatorname{div} \mathbf{f} = \frac{\partial \dot{x}}{\partial x} + \frac{\partial \dot{y}}{\partial y} + \frac{\partial \dot{z}}{\partial z} = 0 + 2y + 0 = 2y
$$

Средняя дивергенция характеризует изменение фазового объема:
- $\langle \operatorname{div} \mathbf{f} \rangle < 0$ — диссипативная система (сжатие фазового объема)
- $\langle \operatorname{div} \mathbf{f} \rangle = 0$ — консервативная система
- $\langle \operatorname{div} \mathbf{f} \rangle > 0$ — расширение фазового объема (неустойчивость)

### 3.4 Размерность Каплана–Йорка

Для хаотических аттракторов вводится дробная размерность Каплана–Йорка:

$$
D_{KY} = j + \frac{\sum_{i=1}^{j} \lambda_i}{|\lambda_{j+1}|}
$$

где $j$ — максимальное целое число, для которого $\sum_{i=1}^{j} \lambda_i > 0$.

Для трехмерных систем ($j=2$):
$$
D_{KY} = 2 + \frac{\lambda_1}{|\lambda_3|}
$$

Эта размерность часто совпадает с информационной размерностью аттрактора и является мерой его фрактальной структуры.

## 4. Сечения Пуанкаре

**Сечение Пуанкаре** — метод редукции размерности системы путем пересечения траекторий с некоторой гиперплоскостью.

Для построения сечения:
1. Выбираем секущую плоскость (например, $z = 0$)
2. Фиксируем моменты пересечения траекторией этой плоскости
3. Откладываем координаты точек пересечения

**Виды сечений:**
- **Двустороннее** — учитываются все пересечения
- **Одностороннее** — учитываются пересечения только в одном направлении (например, $z$ возрастает или убывает)

Сечения Пуанкаре позволяют:
- Визуализировать структуру аттрактора
- Определить тип динамики (периодическая, квазипериодическая, хаотическая)
- Вычислить отображение Пуанкаре

---

## 5. Расхождение близких траекторий

Хаотическая динамика характеризуется экспоненциальным разбеганием близких траекторий. Расстояние между траекториями:

$$
d(t) = \|\mathbf{x}_2(t) - \mathbf{x}_1(t)\| \sim d_0 e^{\lambda_1 t}
$$

На логарифмическом графике это проявляется как линейный участок с наклоном $\lambda_1$.

### 6.2 Спектр показателей Ляпунова

### 6.3 Проверка теоремы о сумме показателей

### 6.4 Размерность Каплана–Йорка

$$
D_{KY} = 2 + \frac{0.1068}{|-0.4958|} = 2 + 0.2154 = 2.2154

Дробная размерность подтверждает фрактальную природу аттрактора.

### 6.5 Характеристики хаоса

- **Время удвоения**: $\tau = \ln 2 / \lambda_1 \approx 6.5$ ед. времени

## 8. Графические результаты

1. **Фазовый портрет** — 3D-траектория с отмеченными стационарными точками
2. **Сечения Пуанкаре** — двустороннее и односторонние сечения плоскостью $z=0$
3. **Расхождение траекторий** — график нормы разности в логарифмическом масштабе
4. **Сходимость показателей Ляпунова** — зависимость $\lambda_i$ от времени интегрирования



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')

# Параметры системы
# dx/dt = 2.7*y + z
# dy/dt = -x + y^2
# dz/dt = x + y

def system(t, state):
    """Исходная система ОДУ"""
    x, y, z = state
    dx = 2.7*y + z
    dy = -x + y**2
    dz = x + y
    return [dx, dy, dz]

def system_with_variations(t, state):

    x, y, z = state[0:3]

    # Исходная система
    dx = 2.7*y + z
    dy = -x + y**2
    dz = x + y

    # Уравнения в вариациях
    # Якобиан: J = [[0, 2.7, 1], [-1, 2y, 0], [1, 1, 0]]
    dvar = []
    for i in range(3):
        dx_var = 0*state[3+3*i] + 2.7*state[4+3*i] + 1*state[5+3*i]
        dy_var = -1*state[3+3*i] + 2*y*state[4+3*i] + 0*state[5+3*i]
        dz_var = 1*state[3+3*i] + 1*state[4+3*i] + 0*state[5+3*i]
        dvar.extend([dx_var, dy_var, dz_var])

    return [dx, dy, dz] + dvar

def gram_schmidt(vectors):
    """Ортогонализация и нормировка векторов по Граму-Шмидту"""
    v = np.array(vectors, dtype=float)
    n = len(v)
    u = np.zeros_like(v)
    norms = np.zeros(n)

    for i in range(n):
        u[i] = v[i].copy()
        for j in range(i):
            u[i] -= np.dot(v[i], u[j]) / np.dot(u[j], u[j]) * u[j]
        norms[i] = np.linalg.norm(u[i])
        if norms[i] > 1e-12:
            u[i] /= norms[i]

    return u, norms

def compute_lyapunov(t_span, y0, dt=0.01, T_step=1.0, n_steps=500):
    """
    Вычисление спектра показателей Ляпунова методом Бенеттина
    """
    n_lyap = 3  # размерность системы
    state = np.zeros(3 + 3*n_lyap)
    state[:3] = y0

    # Начальные векторы возмущений (ортогональные)
    for i in range(n_lyap):
        state[3 + 3*i + i] = 1.0

    t = 0
    sums = np.zeros(n_lyap)

    for step in range(n_steps):
        # Интегрируем на интервале T_step
        # Используем solve_ivp без t_eval для большей точности
        sol = solve_ivp(
            system_with_variations,
            [t, t + T_step],
            state,
            method='RK45',
            rtol=1e-10,
            atol=1e-12,
            max_step=dt  # ограничиваем максимальный шаг
        )

        # Берем последнее состояние
        state = sol.y[:, -1]
        t += T_step

        # Извлекаем векторы возмущений
        vectors = []
        for i in range(n_lyap):
            vec = state[3 + 3*i:3 + 3*(i+1)]
            vectors.append(vec)

        # Ортогонализация и вычисление норм
        _, norms = gram_schmidt(vectors)
        sums += np.log(norms)

        # Обновляем векторы возмущений (уже ортонормированные)
        u, _ = gram_schmidt(vectors)
        for i in range(n_lyap):
            state[3 + 3*i:3 + 3*(i+1)] = u[i]

    lyapunov = sums / (n_steps * T_step)
    return lyapunov

def integrate_trajectory(y0, t_span, method='RK45'):
    """Интегрирование траектории"""
    sol = solve_ivp(system, t_span, y0, method=method, rtol=1e-10, atol=1e-12)
    return sol.t, sol.y

def poincare_section(t, y, plane='z', value=0, direction='both'):
    """
    Построение сечения Пуанкаре
    plane: 'x', 'y', 'z' - фиксируемая переменная
    value: значение плоскости
    direction: 'both', 'positive', 'negative'
    """
    points = []

    for i in range(1, len(t)):
        # Индексы переменных
        idx = {'x': 0, 'y': 1, 'z': 2}[plane]

        if (y[idx, i-1] - value) * (y[idx, i] - value) < 0:
            # Есть пересечение
            sign_prev = np.sign(y[idx, i-1] - value)
            sign_curr = np.sign(y[idx, i] - value)

            if direction == 'both' or (direction == 'positive' and sign_prev < 0 and sign_curr > 0) or \
               (direction == 'negative' and sign_prev > 0 and sign_curr < 0):

                # Линейная интерполяция
                t_intersect = t[i-1] + (t[i] - t[i-1]) * (value - y[idx, i-1]) / (y[idx, i] - y[idx, i-1])

                # Интерполяция других координат
                x_interp = np.interp(t_intersect, t[i-1:i+1], y[0, i-1:i+1])
                y_interp = np.interp(t_intersect, t[i-1:i+1], y[1, i-1:i+1])
                z_interp = np.interp(t_intersect, t[i-1:i+1], y[2, i-1:i+1])

                if plane == 'x':
                    points.append([y_interp, z_interp])
                elif plane == 'y':
                    points.append([x_interp, z_interp])
                else:  # plane == 'z'
                    points.append([x_interp, y_interp])

    return np.array(points)




print("Стационарные точки:")
print("P1 = (0, 0, 0)")
print("P2 = (1, -1, 2.7)")

# Проверка
x, y, z = 0, 0, 0
print(f"\nПроверка P1: dx/dt={2.7*y+z:.2f}, dy/dt={-x+y**2:.2f}, dz/dt={x+y:.2f}")

x, y, z = 1, -1, 2.7
print(f"Проверка P2: dx/dt={2.7*y+z:.2f}, dy/dt={-x+y**2:.2f}, dz/dt={x+y:.2f}")



# Начальные условия
y0 = [0.1, 0.1, 0.1]
t_span = [0, 100]

print("Интегрирование системы.")
t, y = integrate_trajectory(y0, t_span)

# Создание фигуры
fig = plt.figure(figsize=(15, 5))

# 3D фазовый портрет
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot(y[0], y[1], y[2], 'b-', linewidth=0.5)
ax1.scatter([0], [0], [0], c='red', s=100, label='P1 (0,0,0)')
ax1.scatter([1], [-1], [2.7], c='red', s=100, label='P2 (1,-1,2.7)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.set_title('Фазовый портрет (3D)')
ax1.legend()

# Проекция xy
ax2 = fig.add_subplot(132)
ax2.plot(y[0], y[1], 'b-', linewidth=0.5)
ax2.scatter([0], [0], c='red', s=100)
ax2.scatter([1], [-1], c='red', s=100)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Проекция на плоскость xy')
ax2.grid(True)

# Проекция xz
ax3 = fig.add_subplot(133)
ax3.plot(y[0], y[2], 'b-', linewidth=0.5)
ax3.scatter([0], [0], c='red', s=100)
ax3.scatter([1], [2.7], c='red', s=100)
ax3.set_xlabel('x')
ax3.set_ylabel('z')
ax3.set_title('Проекция на плоскость xz')
ax3.grid(True)

plt.tight_layout()
plt.savefig('phase_portrait.png', dpi=150)
plt.show()
print("Фазовый портрет сохранен как 'phase_portrait.png'")




# Длинная траектория для сечения
t_span_long = [0, 10000]
print("Интегрирование длинной траектории (10000 ед. времени)...")
t_long, y_long = integrate_trajectory(y0, t_span_long)

# Сечения плоскостью z = 0
print("Построение сечений плоскостью z=0...")
points_both = poincare_section(t_long, y_long, plane='z', value=0, direction='both')
points_pos = poincare_section(t_long, y_long, plane='z', value=0, direction='positive')
points_neg = poincare_section(t_long, y_long, plane='z', value=0, direction='negative')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Двустороннее сечение
if len(points_both) > 0:
    axes[0].scatter(points_both[:, 0], points_both[:, 1], s=1, alpha=0.5, c='blue')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Двустороннее сечение (z=0)')
axes[0].grid(True)

# Одностороннее (возрастание z)
if len(points_pos) > 0:
    axes[1].scatter(points_pos[:, 0], points_pos[:, 1], s=1, alpha=0.5, c='red')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Одностороннее сечение (z↑ через 0)')
axes[1].grid(True)

# Одностороннее (убывание z)
if len(points_neg) > 0:
    axes[2].scatter(points_neg[:, 0], points_neg[:, 1], s=1, alpha=0.5, c='green')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].set_title('Одностороннее сечение (z↓ через 0)')
axes[2].grid(True)

plt.tight_layout()
plt.savefig('poincare_sections.png', dpi=150)
plt.show()
print("Сечения Пуанкаре сохранены как 'poincare_sections.png'")


# Две близкие траектории
y0_1 = [0.1, 0.1, 0.1]
y0_2 = [0.1 + 1e-6, 0.1, 0.1]

t_span_div = [0, 50]
t_div, y_div1 = integrate_trajectory(y0_1, t_span_div, method='RK45')
_, y_div2 = integrate_trajectory(y0_2, t_span_div, method='RK45')

# Вычисление расстояния
distance = np.sqrt(np.sum((y_div2 - y_div1)**2, axis=0))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Разность траекторий (модуль)
ax1.semilogy(t_div, distance, 'b-', linewidth=1)
ax1.set_xlabel('Время t')
ax1.set_ylabel('Расстояние ||Δr||')
ax1.set_title('Расхождение двух близких траекторий (логарифмический масштаб)')
ax1.grid(True)

# Для наглядности - первые 10 секунд
mask = t_div < 10
ax2.semilogy(t_div[mask], distance[mask], 'r-', linewidth=1.5)
ax2.set_xlabel('Время t')
ax2.set_ylabel('Расстояние ||Δr||')
ax2.set_title('Расхождение траекторий (первые 10 секунд)')
ax2.grid(True)

plt.tight_layout()
plt.savefig('trajectory_divergence.png', dpi=150)
plt.show()
print("График расхождения сохранен как 'trajectory_divergence.png'")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')

def system(t, state):
    """Исходная система ОДУ"""
    x, y, z = state
    dx = 2.7*y + z
    dy = -x + y**2
    dz = x + y
    return [dx, dy, dz]

def system_with_variations(t, state):
    """Расширенная система: исходная + 3 набора уравнений в вариациях"""
    x, y, z = state[0:3]

    # Исходная система
    dx = 2.7*y + z
    dy = -x + y**2
    dz = x + y

    # Якобиан: J = [[0, 2.7, 1], [-1, 2y, 0], [1, 1, 0]]
    dvar = []
    for i in range(3):
        dx_var = 0*state[3+3*i] + 2.7*state[4+3*i] + 1*state[5+3*i]
        dy_var = -1*state[3+3*i] + 2*y*state[4+3*i] + 0*state[5+3*i]
        dz_var = 1*state[3+3*i] + 1*state[4+3*i] + 0*state[5+3*i]
        dvar.extend([dx_var, dy_var, dz_var])

    return [dx, dy, dz] + dvar

def gram_schmidt(vectors):
    """Ортогонализация и нормировка векторов по Граму-Шмидту"""
    v = np.array(vectors, dtype=float)
    n = len(v)
    u = np.zeros_like(v)
    norms = np.zeros(n)

    for i in range(n):
        u[i] = v[i].copy()
        for j in range(i):
            u[i] -= np.dot(v[i], u[j]) / np.dot(u[j], u[j]) * u[j]
        norms[i] = np.linalg.norm(u[i])
        if norms[i] > 1e-12:
            u[i] /= norms[i]

    return u, norms

def compute_lyapunov_accurate(y0, t_total=1000, T_step=0.5, n_steps=2000):
    """
    Улучшенное вычисление спектра показателей Ляпунова
    """
    n_lyap = 3
    state = np.zeros(3 + 3*n_lyap)
    state[:3] = y0

    # Начальные ортонормированные векторы возмущений
    for i in range(n_lyap):
        state[3 + 3*i + i] = 1.0

    t = 0
    sums = np.zeros(n_lyap)
    step_count = 0

    # Сначала даем системе выйти на аттрактор

    transient = 100
    sol = solve_ivp(
        system_with_variations,
        [0, transient],
        state,
        method='RK45',
        rtol=1e-10,
        atol=1e-12
    )
    state = sol.y[:, -1]
    t = transient

    print(f"Вычисление показателей Ляпунова ({n_steps} шагов)")
    for step in range(n_steps):
        # Интегрируем на интервале T_step
        sol = solve_ivp(
            system_with_variations,
            [t, t + T_step],
            state,
            method='DOP853',  # Более точный метод
            rtol=1e-12,
            atol=1e-14,
            max_step=0.01
        )

        state = sol.y[:, -1]
        t += T_step

        # Извлекаем векторы возмущений
        vectors = []
        for i in range(n_lyap):
            vec = state[3 + 3*i:3 + 3*(i+1)]
            vectors.append(vec)

        # Ортогонализация и вычисление норм
        _, norms = gram_schmidt(vectors)
        sums += np.log(norms)
        step_count += 1

        # Обновляем векторы возмущений
        u, _ = gram_schmidt(vectors)
        for i in range(n_lyap):
            state[3 + 3*i:3 + 3*(i+1)] = u[i]

    lyapunov = sums / (step_count * T_step)
    return lyapunov

def compute_average_divergence(y0, t_span=[0, 1000], dt=0.01):
    """
    Вычисление средней дивергенции вдоль траектории
    """
    # Интегрируем траекторию
    t_eval = np.arange(t_span[0], t_span[1], dt)
    sol = solve_ivp(
        system,
        t_span,
        y0,
        method='DOP853',
        t_eval=t_eval,
        rtol=1e-12,
        atol=1e-14
    )

    # Вычисляем дивергенцию в каждой точке
    divergence_values = []
    for i in range(len(sol.t)):
        x, y, z = sol.y[:, i]
        # div F = ∂(2.7y+z)/∂x + ∂(-x+y²)/∂y + ∂(x+y)/∂z = 0 + 2y + 0
        div = 2*y
        divergence_values.append(div)

    mean_div = np.mean(divergence_values)
    return mean_div, sol.t, divergence_values


# Начальные условия
y0 = [0.1, 0.1, 0.1]


lyapunov = compute_lyapunov_accurate(y0, t_total=1000, T_step=0.5, n_steps=2000)


mean_div, t_div, div_values = compute_average_divergence(y0, t_span=[0, 1000], dt=0.01)


print(f"\nПоказатели Ляпунова:")
print(f"λ₁ = {lyapunov[0]:.8f}")
print(f"λ₂ = {lyapunov[1]:.8f}")
print(f"λ₃ = {lyapunov[2]:.8f}")
print(f"Сумма λ = {np.sum(lyapunov):.8f}")

print(f"\nСредняя дивергенция:")
print(f"⟨div F⟩ = {mean_div:.8f}")

print(f"\nПроверка:")
print(f"Разница = {np.abs(np.sum(lyapunov) - mean_div):.8f}")
print(f"Относительная погрешность = {np.abs(np.sum(lyapunov) - mean_div)/np.abs(mean_div)*100:.2f}%")



# Проверка сходимости с разным временем интегрирования
T_values = [100, 200, 500, 1000, 2000]
lyapunov_vs_time = []

print("\nИсследование зависимости от времени интегрирования:")
for T in T_values:
    print(f"  Время интегрирования: {T}")
    lyap = compute_lyapunov_accurate(y0, t_total=T, T_step=0.5, n_steps=int(T/0.5))
    lyapunov_vs_time.append(lyap)

# Построение графиков
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# График сходимости показателей Ляпунова
ax1 = axes[0, 0]
for i in range(3):
    vals = [l[i] for l in lyapunov_vs_time]
    ax1.plot(T_values, vals, 'o-', label=f'λ_{i+1}', markersize=8)
ax1.axhline(y=0, color='k', linestyle='--', alpha=0.5)
ax1.set_xlabel('Время интегрирования', fontsize=12)
ax1.set_ylabel('Показатели Ляпунова', fontsize=12)
ax1.set_title('Сходимость показателей Ляпунова', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

# График дивергенции
ax2 = axes[0, 1]
# Сглаживаем для наглядности
window = 100
smooth_div = np.convolve(div_values, np.ones(window)/window, mode='valid')
t_smooth = t_div[window//2:-window//2+1]
ax2.plot(t_smooth, smooth_div, 'b-', linewidth=1, alpha=0.7)
ax2.axhline(y=mean_div, color='r', linestyle='--', label=f'Среднее = {mean_div:.4f}')
ax2.set_xlabel('Время', fontsize=12)
ax2.set_ylabel('Дивергенция', fontsize=12)
ax2.set_title('Дивергенция вдоль траектории', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)

# График разности
ax3 = axes[1, 0]
diff = [np.sum(l) for l in lyapunov_vs_time]
ax3.plot(T_values, diff, 'g-o', markersize=8, label='Сумма λ')
ax3.axhline(y=mean_div, color='r', linestyle='--', label=f'⟨div F⟩ = {mean_div:.4f}')
ax3.set_xlabel('Время интегрирования', fontsize=12)
ax3.set_ylabel('Значение', fontsize=12)
ax3.set_title('Сумма показателей Ляпунова vs средняя дивергенция', fontsize=14)
ax3.legend()
ax3.grid(True, alpha=0.3)

# Гистограмма распределения дивергенции
ax4 = axes[1, 1]
ax4.hist(div_values, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')
ax4.axvline(x=mean_div, color='r', linestyle='--', linewidth=2, label=f'Среднее = {mean_div:.4f}')
ax4.set_xlabel('Дивергенция', fontsize=12)
ax4.set_ylabel('Плотность вероятности', fontsize=12)
ax4.set_title('Распределение дивергенции', fontsize=14)
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('lyapunov_analysis.png', dpi=150)
plt.show()


if lyapunov[0] > 0:
    print("\n λ₁ > 0 → система имеет чувствительную зависимость от начальных условий")
    print("  Аттрактор является ХАОТИЧЕСКИМ")
    print(f"  Характерное время удвоения: τ = ln(2)/λ₁ = {np.log(2)/lyapunov[0]:.2f} ед. времени")
else:
    print("\n λ₁ ≤ 0 → хаотическое поведение не обнаружено")

if np.abs(lyapunov[1]) < 0.01:
    print(" λ₂ ≈ 0 → направление, соответствующее движению вдоль траектории")

if lyapunov[2] < 0:
    print(" λ₃ < 0 → сжатие в одном направлении обеспечивает устойчивость")

print(f"\nСпектр показателей Ляпунова:")
print(f"  λ₁ = {lyapunov[0]:.6f} (растяжение)")
print(f"  λ₂ = {lyapunov[1]:.6f} (нулевое)")
print(f"  λ₃ = {lyapunov[2]:.6f} (сжатие)")

# Размерность Каплана-Йорка
if lyapunov[2] != 0:
    dky = 2 + lyapunov[0]/np.abs(lyapunov[2])
    print(f"\nРазмерность Каплана-Йорка: D_KY = {dky:.4f}")
    print("Это дробная размерность, характерная для странного аттрактора")

# Проверка на диссипативность
print(f"\nСистема является ДИССИПАТИВНОЙ" if mean_div < 0 else "Система является КОНСЕРВАТИВНОЙ")
print(f"Среднее сжатие фазового объема: {mean_div:.6f}")

```
